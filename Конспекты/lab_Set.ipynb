# Изменяемое множество (set)

`Изменяемое множетсво (set)` - неупорядоченная изменяемая структура данных с уникальными элементами.

```python
{2,3,3,4, "hello"}
```

```python
set('hello')
```

```python
cities = "Воронеж, Воронеж, Тула, Курск, Курс".split(", ")
print(cities)
cities_un= list(set(cities))
print(cities_un)
```
