```python
url_base = 'https://static-maps.yandex.ru/1.x'
url_params = dict ()
url_params['l'] = 'map' # схема
url_params['ll'] = "37.694313,55.751068" # центр карты
url_params['z'] = 12
print(url_params)
```

```python
import requests
from PIL import Image
```

```python
url_params['1'] = 'map,skl,rtf'
# map - схема
# trf - траффик
# sat - спутник
# skl- название
url_params['z']=15
url_params['pt'] = '37.481421,55.836372' #pm2a1 - не работает
with requests.get(url=url_base, params=url_params,stream=True) as r:
    im = Image.open(r.raw)
print(im.format)
print(im.mode)
print(im.size)
im
```
