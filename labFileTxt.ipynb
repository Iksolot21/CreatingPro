# Работа с текстовыми файлами

Функция `open()` - открыть файл на чтение/запись.

Режимы открытия файла:
- `'r'` - открыть файл для чтения. Если файла нет, то ошибка. (по умолчанию)
- `'w'` - открыть файл для перезаписи. Если файла нет то создать.
- `'x'` - открыть файл для перезаписи. Если файла нет то создать.
- `'a'` - открыть файл для добавления в конец. Если файла нет то создать.

Дополнительные режимы:
- `'t'` - Текстовый файл (по умолчанию)
- `'b'` - Бинарный файл 
- `'+'` - Режим R + W 

Помним:
- файл необходимо закрыть.
- различные кодировки текстовых файлов.

```python
import os
TEMP_DIR = 'temp'
if not os.path.exists(TEMP_DIR):
    os.mkdir(TEMP_DIR)
```

```python
os.path.join(TEMP_DIR, 'file_TXT.txt')
```

```python
f = open(os.path.join(TEMP_DIR, 'file_TXT.txt'), 'w')
try:
    f.write('Hello\n')
    f.write('1234\n')   
finally:
    f.close()
```

```python
with open(os.path.join(TEMP_DIR, 'file_TXT.txt'), 'w'):
    f.write('Hello\n')
    f.write('1234\n')   
```

```python
poem='''Духовной жаждою томим,
В пустыне мрачной я влачился,
И шестикрылый серафим
На перепутье мне явился;
Перстами легкими как сон
Моих зениц коснулся он:
Отверзлись вещие зеницы,
Как у испуганной орлицы'''
```

```python

with open(os.path.join(TEMP_DIR, 'file_TXT_POEM.txt'), 'w') as f:
    f.write(poem)
```

```python
with open(os.path.join(TEMP_DIR, 'file_TXT_POEM.txt'), 'w', endoding = 'utf-8') as f:
    f.write(poem)
```

# Чтение из файла

```python
with open(os.path.join(TEMP_DIR, 'file_TXT_POEM.txt')) as f:
    print(f.read(13))
```

```python
with open(os.path.join(TEMP_DIR, 'file_TXT_POEM.txt')) as f:
    for line in f:
        line=line.rstrip('\n')
        print(repr(line))
```

```python
with open(os.path.join(TEMP_DIR, 'file_TXT_POEM.txt')) as f:
    print(f.readlines())
```

```python
with open(os.path.join(TEMP_DIR, 'file_TXT_POEM.txt')) as f:
        while True: 
                if line:= f.readlines() == "":
                    break
                print(repr(line))
```

```python
with open(os.path.join(TEMP_DIR, 'file_TXT_POEM.txt')) as f:
        while line:= f.readlines(): 
                print(repr(line))
```
