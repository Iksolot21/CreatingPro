# Пакет Pillow

* Pillow - это форк PIL.

```python
from PIL import Image
```

```python
Image.__version__ 
```

```python
im = Image.open('image.jpg')
# display(im)
im
```

```python
print(im.format)
print(im.format_description)
print(im.mode)
print(im.size)
print(im.height)
print(im.width)
print(im.getexif())
```

```python
im.info
```

```python
# обрезать
im.crop((50,50,500,200))
```

```python
# преобразовать в миниатюру с сохранением пропорций
imCopy = im.copy()
imCopy.thumbnail((200,200))
imCopy
```

```python
# без сохранения пропорций
im.resize((200,200))
```

```python
im = im.resize((im.width //4, im.height//4))
```

## Транспорирование изображение

```python
im.transpose(Image.Transpose.FLIP_LEFT_RIGHT)
```

```python
im.transpose(Image.Transpose.ROTATE_90)
```

## Свободное вращение

```python
im.rotate(45)
```

```python
im.rotate(45,expand=True)
```

```python
im.rotate(45, center=(im.width/2, im.height))
```

```python
im.rotate(45, fillcolor=(213,125,90))
```
