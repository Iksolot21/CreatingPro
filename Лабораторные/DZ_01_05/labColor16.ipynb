```python
# Коды цветов для терминала
text_color_codes = [30, 31, 32, 33, 34, 35, 36, 37]
background_color_codes = [40, 41, 42, 43, 44, 45, 46, 47]

# Вывод всех цветов текста на всех фонах
print("Цвета текста на разных фонах:")
for text_code in text_color_codes:
    for background_code in background_color_codes:
        color_code = f"\033[{text_code};{background_code}m"
        reset_code = "\033[0m"
        print(f"{color_code}{text_code};{background_code}{reset_code}", end=" | ")
    print()
print()

# Вывод всех фонов с разными цветами текста
print("Цвета фона с разными цветами текста:")
for background_code in background_color_codes:
    for text_code in text_color_codes:
        color_code = f"\033[{text_code};{background_code}m"
        reset_code = "\033[0m"
        print(f"{color_code}{text_code};{background_code}{reset_code}", end=" | ")
    print()
print()
```
