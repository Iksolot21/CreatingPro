```python
from PIL import Image, ImageDraw
from math import cos, sin, radians

def draw_star(draw, center, size, color):
    x, y = center
    points = [
        (x, y - size),  # Top
        (x + size // 3, y - size // 3),
        (x + size, y - size // 3),
        (x + size // 2, y + size // 6),
        (x + 3 * size // 4, y + size),
        (x, y + size // 2),
        (x - 3 * size // 4, y + size),
        (x - size // 2, y + size // 6),
        (x - size, y - size // 3),
        (x - size // 3, y - size // 3)
    ]
    draw.polygon(points, outline=color)

def draw_hexagon(draw, center, size, color):
    x, y = center
    points = [
        (x + size, y),
        (x + size // 2, y + int(size * 0.866)),
        (x - size // 2, y + int(size * 0.866)),
        (x - size, y),
        (x - size // 2, y - int(size * 0.866)),
        (x + size // 2, y - int(size * 0.866))
    ]
    draw.polygon(points, outline=color)

def draw_starburst(draw, center, size, color, num_lines=36):
    x, y = center
    for i in range(num_lines):
        angle = i * 360 / num_lines
        x_end = x + size * cos(radians(angle))
        y_end = y + size * sin(radians(angle))
        draw.line([(x, y), (x_end, y_end)], fill=color)

def draw_gradient(draw, top_left, bottom_right, start_color, end_color):
    x0, y0 = top_left
    x1, y1 = bottom_right
    r0, g0, b0 = start_color
    r1, g1, b1 = end_color
    for y in range(y0, y1 + 1):
        ratio = (y - y0) / (y1 - y0)
        r = int(r0 + (r1 - r0) * ratio)
        g = int(g0 + (g1 - g0) * ratio)
        b = int(b0 + (b1 - b0) * ratio)
        draw.line([(x0, y), (x1, y)], fill=(r, g, b))

# Пример использования функций
img_size = 400
image = Image.new('RGB', (img_size, img_size), (255, 255, 255))
draw = ImageDraw.Draw(image)

# Рисование фигур
draw_star(draw, (100, 100), 50, 'black')
draw_hexagon(draw, (300, 100), 50, 'black')
draw_starburst(draw, (100, 300), 100, 'black')
draw_gradient(draw, (200, 200), (399, 399), (0, 0, 0), (255, 255, 255))

image.show()
```
