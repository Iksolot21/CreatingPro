```python
from labToolsHashFilesStat import HashTimeTable

# Получить статистику времени для всех доступных алгоритмов хеширования в текущей директории и ее подкаталогах
time_stats = HashTimeTable('.', recursive=True)

# Вывести таблицу со статистикой времени в консоль, отсортированную по среднему времени (от меньшего к большему)
for hash_name, stats in sorted(time_stats.items(), key=lambda x: x[1]['average']):
    print(f"{hash_name}:")
    print(f"  Минимальное время: {stats['min']:.6f} с")
    print(f"  Максимальное время: {stats['max']:.6f} с")
    print(f"  Среднее время: {stats['average']:.6f} с")
    print(f"  Суммарное время: {stats['total']:.6f} с")
    print()
```
