```python
from labToolsHashFilesStat import HashTimeTable, format_time

time_stats = HashTimeTable('.', recursive=True)

# Вывести таблицу со статистикой времени в консоль, отсортированную по среднему времени (от меньшего к большему)
for hash_name, stats in sorted(time_stats.items(), key=lambda x: x[1]['average']):
    print(f"{hash_name}:")
    print(f"  Минимальное время: {format_time(stats['min'])}")
    print(f"  Максимальное время: {format_time(stats['max'])}")
    print(f"  Среднее время: {format_time(stats['average'])}")
    print(f"  Суммарное время: {format_time(stats['total'])}")
    print()
```
