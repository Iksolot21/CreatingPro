# “labGenComprehension” - Генераторы списков
Создать коллекции.
- Из строки:
  - в каждом слове поменять регистр символов на противоположный
  - в каждом слове заменить каждую букву на '_'
  - в каждом слове первую и последнюю букву заменить на '_'
  - в каждом слове заменить каждую букву на '_', если длина слова меньше 5 символов
  - в каждом слове поменять буквы местами случайным образом
  - каждую последнюю букву в слове написать с большой буквы
  - каждое второе слово написать большими буквами, а остальные слова маленькими
  - поменять слова местами случайным образом
  - сформировать список слов, отсортированных по количеству букв в каждом слове
  - сформировать список уникальных букв
  - сформировать список уникальных букв и их количество
  - ...
- придумать свои примеры по генерации списков/множеств/словарей

```python
text = "На солнечном небе плывут облака, будто БЕлЫЕ птицы в бескрайнем просторе. Ветер играет с листьями."
print([x.lower() if x.isupper() else x.upper() for i in text.split() for x in i])
print(["_" if x.isalpha() else x for i in text.split() for x in i])
print(['_' if (l==0 or l==len(i)-1) else x for i in text.split() for l,x in enumerate(i)])
print(["_" if len(i)<5 else x for i in text.split() for x in i])
import random
print([''.join(random.sample(i, len(i))) for i in text.split()])
print([ x.upper() if (l==len(i)-1) else x for i in text.split() for l,x in enumerate(i)])
print([i.upper() if l % 2 == 0 else i.lower() for l, i in enumerate(text.split())])
print(sorted(text.split(), key=len))
print({char.lower(): text.count(char) for char in text if char.isalpha()})
# свои примеры 
print({char: ord(char) for char in text})
print([len(word) for word in text.split()])







```
